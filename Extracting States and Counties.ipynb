```python
import pandas as pd
import re


# Initialize empty lists to store state codes and countries
state_codes = []
countries = []

# Regular expressions for state codes and countries
state_pattern = r'\b([A-Z]{2})\b'
country_pattern = r',\s*([A-Z][a-z\s]+)\b'

# Iterate through the series and extract state codes and countries
for description in df.name:
    state_match = re.search(state_pattern, description)
    country_match = re.search(country_pattern, description)
    
    if state_match:
        state_codes.append(state_match.group(1))
    else:
        state_codes.append(None)  # If no state code is found, use None
    
    if country_match:
        countries.append(country_match.group(1))
    else:
        countries.append(None)  # If no country is found, use None

# Add state codes and countries as new columns to the DataFrame
df['State_Code'] = state_codes
df['Country'] = countries

# Display the DataFrame with extracted data
print(df)
```

```python
import geocoder

# Create 'state' and 'country' columns with empty values
df['state'] = ''
df['country'] = ''

# Iterate over each row in the DataFrame
record_len = len(df)  # Get the total number of records for progress tracking
for index, (lat, lon) in enumerate(zip(df.lat, df.lon)):
    try:
        # Perform reverse geocoding
        location = geocoder.osm([lat, lon], method='reverse')

        if location:
            state_code = location.state
            place = location.city
            country = location.country
            # Update the 'state' and 'country' columns in the DataFrame
            df.at[index, 'state'] = state_code
            df.at[index, 'country'] = country
            print(f"State: {state_code}")
            print(f"Place: {place}")
        else:
            print("Location not found")

    except Exception as e:
        print(f"Error processing record at index {index}: {str(e)}")

    # Print progress
    print("Percentage completed:", int((index + 1) * 100 / record_len), '%')

```
